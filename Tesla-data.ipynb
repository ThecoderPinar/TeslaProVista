```python
import pandas as pd
```

```python
hisse_senedi_data = pd.read_csv('Tesla-dataset.csv')
```

```python
hisse_senedi_data.dropna(inplace=True)  # Eksik verileri çıkarabilirsiniz veya fillna ile doldurabilirsiniz
```

```python
hisse_senedi_data["Date"] = pd.to_datetime(hisse_senedi_data["Date"])
```

```python
hisse_senedi_data.set_index("Date", inplace=True)
```

```python
import matplotlib.pyplot as plt
%pip install plotly
```

```python
import plotly.express as px


fig = px.line(hisse_senedi_data, x="Date", y="Close", title="Tesla Stock Closing Prices")
fig.update_traces(line=dict(color="green"))
fig.update_layout(
xaxis=dict(title="Date"),
yaxis=dict(title="Sales"),
hovermode="x unified",  # Etkileşimli hoover özelliği
template="plotly_dark"  # Tema seçimi
)
fig.show()
```

```python
hisse_senedi_data["Price Change"] = hisse_senedi_data["Close"].diff()

```

```python
daily_returns = hisse_senedi_data["Price Change"].dropna()
```

```python
import numpy as np

std_dev = np.std(daily_returns)
```

```python
confidence_levels = [0.90, 0.95, 0.99]
```

```python

var_5_percentile = -0.050963512987485034
var_95_percentile = 0.05387099692277739
```

```python
import plotly.graph_objects as go

# VaR analizi grafiği oluşturma
fig = go.Figure()

fig.add_trace(go.Bar(
    x=["5% Quantile", "95% Quantile"],
    y=[var_5_percentile, var_95_percentile],
    text=[f"{var_5_percentile:.4f}", f"{var_95_percentile:.4f}"],
    textposition="auto",
    marker_color=["red", "green"]
))

fig.update_layout(
    title="Value at Risk (VaR) Analysis",
    xaxis_title="Quantile",
    yaxis_title="VaR Value",
    template="plotly_dark",
    showlegend=False
)

fig.update_traces(
    hovertemplate="<b>%{x}</b><br>VaR: %{text}",
)

fig.show()
```

# Value at Risk (VaR) Analysis

This graph represents the Value at Risk (VaR) analysis for Tesla's stock based on calculated quantiles.

- The red bar represents the 5% quantile, showing the potential worst-case scenario for a single day.
- The green bar represents the 95% quantile, indicating a more optimistic scenario for a single day.

VaR values are calculated using statistical methods, giving insights into potential losses at different confidence levels.

For Tesla's stock, the calculated VaR values are as follows:
- 5% Quantile: -0.05096
- 95% Quantile: 0.05387
